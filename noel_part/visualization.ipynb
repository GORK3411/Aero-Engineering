```python
import os
import pandas as pd
from traffic.core import Traffic, Flight
import webbrowser
cleaned_data_folder = "cleaned_data_final"
dirs = [d for d in os.listdir(cleaned_data_folder) if os.path.isdir(os.path.join(cleaned_data_folder, d))]
for d in dirs:
    print(d)
```

```python
step = dirs[0]
```

```python
cur_directory = os.path.join(cleaned_data_folder, step)
flights = [f for f in os.listdir(cur_directory)]
for f in flights:
    print(f)
```

```python
selected_flight = flights[0]
```

```python
step = "step1_raw_2023-08-10_to_2023-09-10"
selected_flight = "20230816_7380c1_065449_081655"
```

```python
reconstructed_path = "final_reconstructions"
df_ads_b_before = pd.read_parquet(rf"{cleaned_data_folder}\\{step}\\{selected_flight}\\adsb_before.parquet")
df_ads_c = pd.read_parquet(rf"{cleaned_data_folder}\\{step}\\{selected_flight}\\adsc.parquet")
df_ads_b_after = pd.read_parquet(rf"{cleaned_data_folder}\\{step}\\{selected_flight}\\adsb_after.parquet")
df_b_c = pd.read_parquet(rf"{reconstructed_path}\\{step}\\{selected_flight}\\adsb_before-adsc.parquet")  
df_c_b = pd.read_parquet(rf"{reconstructed_path}\\{step}\\{selected_flight}\\adsc-adsb_after.parquet")
df_b_b = pd.read_parquet(rf"{reconstructed_path}\\{step}\\{selected_flight}\\adsb_before-adsb_after.parquet")
df_final = pd.read_parquet(rf"{reconstructed_path}\\{step}\\{selected_flight}\\full_reconstruction.parquet")
flight_ads_b_before = Flight(df_ads_b_before)
flight_ads_c = Flight(df_ads_c)
flight_ads_b_after = Flight(df_ads_b_after)
flight_b_c = Flight(df_b_c)
flight_c_b = Flight(df_c_b)
flight_b_b = Flight(df_b_b)
final_flight = Flight(df_final)
flight_ads_b_before | flight_ads_c | flight_ads_b_after | flight_b_c | flight_c_b | flight_b_b | final_flight
```

```python
import pandas as pd
from traffic.core import Traffic

dfs = []
# to visualize multiple flights on map put them in the list here, they are all in same color
for i, flight in enumerate([flight_ads_b_before, flight_ads_c, flight_ads_b_after]):
    df = flight.data.copy()
    df["flight_id"] = f"segment_{i}"
    dfs.append(df)

combined = pd.concat(dfs, ignore_index=True)
t = Traffic(combined, flight_id="flight_id")
display(t.map_leaflet())
m = t.map_leaflet()
m.save(f"{reconstructed_path}\\{step}\\{selected_flight}\\original_segments.html")
webbrowser.open(f"{reconstructed_path}\\{step}\\{selected_flight}\\original_segments.html")
```

```python
#single flight visualization
final_map = final_flight.map_leaflet()
display(final_map)
final_map.save(f"{reconstructed_path}\\{step}\\{selected_flight}\\final_map.html")
webbrowser.open(f"{reconstructed_path}\\{step}\\{selected_flight}\\final_map.html")
```
